```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Occupation data
occupations= pd.read_csv('Data/occupations.csv')

print(occupations.shape)
occupations.head()
```

```python
# Load Relationship data
relationships = pd.read_csv('Data/occupationSkillRelations.csv')

print(relationships.shape)
relationships.head()
```

```python
# Load Skills data
skills = pd.read_csv('Data/skills.csv')

print(skills.shape)
skills.head()
```

```python
# Merging occupations and relationships data on occupationUri
occupation_relations= pd.merge(occupations, relationships, on='occupationUri')

print(occupation_relations.shape)
occupation_relations.head()
```

```python
# List all Columns
occupation_relations.columns
```

```python
# Drop Columns
occupation_relations.drop(['conceptType', 'iscoGroup', 'hiddenLabels', 'status', 'modifiedDate', 
                            'regulatedProfessionNote','scopeNote','definition',
                           'inScheme'], axis=1, inplace=True)

print(occupation_relations.shape)
occupation_relations.head()
```

```python
# Merging occupation_relations and skills data on skillUri
occupation_skills= pd.merge(occupation_relations, skills, on='skillUri')

print(occupation_skills.shape)
occupation_skills.head()
```

```python
# List all Columns
occupation_skills.columns
```

```python
# Drop Columns
occupation_skills.drop(['code', 'conceptType', 'skillType_y', 'reuseLevel',
                     'hiddenLabels', 'status', 'modifiedDate', 'scopeNote',
                     'definition', 'inScheme', ], axis=1, inplace=True)

print(occupation_skills.shape)
occupation_skills.head()
```

```python
# Find Nan Values
occupation_skills.isnull().sum()
```

```python
# Rename Columns
occupation_skills.rename(columns={'skillType_x':'skillType'}, inplace=True)
occupation_skills.rename(columns={'preferredLabel_x':'occupationLabel'}, inplace=True)
occupation_skills.rename(columns={'preferredLabel_y':'skillLabel'}, inplace=True)
occupation_skills.rename(columns={'altLabels_x':'alt_occupationLabel'}, inplace=True)
occupation_skills.rename(columns={'altLabels_y':'alt_skillLabel'}, inplace=True)
occupation_skills.rename(columns={'description_x':'occupation_descritption'}, inplace=True)
occupation_skills.rename(columns={'description_y':'skill_descritption'}, inplace=True)
```

```python
occupation_skills.columns
```

```python
#Drop the URI columns
occupation_skills.drop(['occupationUri', 'skillUri'], axis=1, inplace=True)
```

```python
# Save the data
occupation_skills.to_csv('Data/occupation_skills.csv', index=False)
```
